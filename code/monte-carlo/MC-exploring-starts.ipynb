```python
import numpy as np
import sys

if "../" not in sys.path:
  sys.path.append("../") 
from env.blackjack import BlackjackEnv
```

```python
env = BlackjackEnv()

# Initial value function
Q = {}
Returns = {}
Counts = {}

policy = {}

for score in range(11,22):
    for usable_ace in range(2):
        for dealer_score in range(1, 11):

            state = (score, dealer_score, bool(usable_ace))
            
            # Initialize Q
            Q[(state, 0)] = 0
            Q[(state, 1)] = 0
            
            # Initial policy
            if score == 20 or score == 21:
                policy[state] = 0
            else:
                policy[state] = 1
                
```

```python
def gen_episode():
    episode = []
    
    # Random initial state
    score = np.random.randint(11, 22)
    usable_ace = bool(np.random.randint(0, 2))
    dealer_score = np.random.randint(1, 11)
    current_state = (score, dealer_score, usable_ace)
    
    while(True):
        action = policy[current_state] 
        next_state, reward, done, _ = env.step(action)
        episode.append((current_state, action, reward))
        if done:
            break
        current_state = next_state
        
    return episode
```

```python
n_episodes = 10000

for _ in range(n_episodes):
    
    episode = gen_episode()
    
    # State-action pairs in episode
    sa_in_episode = [(sar[0], sar[1]) for sar in episode]    
    
    for i,(s,a) in enumerate(sa_in_episode):
        
        # First-visit only
        if i==0 or (i>=1 and (s,a) not in sa_in_episode[:i-1]):
            
            # Sum up returns (no discount factor)
            G = sum([sar[2] for i,sar in enumerate(episode[i:]) if sar[0]==s and sar[1]==a])
            
            if (s,a) in Returns.keys():
                Returns[(s,a)] += G
            else:
                Returns[(s,a)] = G
                
            if (s,a) in Counts.keys():
                Counts[(s,a)] += 1.0
            else:
                Counts[(s,a)] = 1.0
            
            Q[(s,a)] = Returns[(s,a)]/Counts[(s,a)]
            
            should_hit = 0
            if Q[(s, 1)] > Q[(s, 0)]:
                should_hit = 1

            policy[s] = should_hit
policy
```
